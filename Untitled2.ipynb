```python
def slice_sine(fi_sample, period = 2000):
    sine_collected = []
    detection_thr = 0.02                     ## Collect the sine only when the difference between max and min is larger than this vaue
    total_index = fi_sample.shape[0]
    print("total_index = ", total_index)
 
    cur_index = 0

    while cur_index < total_index-period:
        
        min_val_index = np.argmin(fi_sample[cur_index:cur_index+period])
        min_val = np.min(fi_sample[cur_index:cur_index+period])
        cur_index = min_val_index + cur_index
       

        max_val = np.max(fi_sample[cur_index:cur_index+period])
        min_val = np.min(fi_sample[cur_index:cur_index+period])

        

 

        if (max_val - min_val) > detection_thr: 
            ## To ensure only sine-like functions are collected
            if (fi_sample[cur_index+period//2] > fi_sample[cur_index]) & (fi_sample[cur_index+period//2] > fi_sample[cur_index+period]):
                ## 상승하는 곡선이어야 함
                if (fi_sample[cur_index] < fi_sample[cur_index+1]) & (fi_sample[cur_index+1] < fi_sample[cur_index+2]):
                    sine_collected.append((fi_sample[cur_index:cur_index+period] - min_val)/(max_val-min_val))
            cur_index = cur_index + period * 2
        else:
            cur_index = cur_index + period * 2
            min_val_index = np.argmin(fi_sample[cur_index:cur_index+period])
            min_val = np.min(fi_sample[cur_index:cur_index+period])
    return sine_collected
```

```python
def byte_to_float(fi, num_bytes):
    LP_vector = []
    i = 0
    while i < num_bytes:
        data = fi.read(4)
        i += 4
        unpacked = struct.unpack("<f", data)
        LP_vector.append(np.abs(unpacked))
    return np.array(LP_vector)
```
